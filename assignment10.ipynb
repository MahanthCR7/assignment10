1. JavaScript is Single-Threaded (Blocking vs Async)
Question:

Show two competing tasks: one blocking and one async.

Answer:

```python
function block() {
    let start = Date.now();
    while (Date.now() - start < 3000) {}
    console.log("Blocking task done");
}

function asyncTask() {
    Promise.resolve().then(() => {
        console.log("Async task done");
    });
}

block();
asyncTask();
console.log("End");
```

2. Why JavaScript is Not Async by Default
Explanation:

JavaScript is synchronous by default because it executes code line by line on a single thread. Async behavior is handled by browser APIs and the event loop, not the core language.

```python
console.log("Start");

function test() {
    for (let i = 0; i < 3; i++) {
        console.log(i);
    }
}

test();

console.log("End");
```

3. Chaining Promises with setTimeout
Question:

Chain promises to print messages with delay.

Answer:

```python
function delay(msg, time) {
    return new Promise(resolve => {
        setTimeout(() => {
            console.log(msg);
            resolve();
        }, time);
    });
}

delay("First", 1000)
.then(() => delay("Second", 1000))
.then(() => delay("Third", 1000));
```

4. States of a Promise

Transition:
Pending → Fulfilled or Rejected (once changed, cannot go back)

5. Event Loop: Promises vs setTimeout
Answer:
Promises go to Microtask Queue
setTimeout goes to Callback Queue (Macrotask Queue)
Event loop executes:
Call stack
Microtask queue (Promises)
Callback queue (setTimeout)

So Promises execute before setTimeout even if delay is 0.

```python
console.log("Start");

setTimeout(() => {
    console.log("setTimeout");
}, 0);

Promise.resolve().then(() => {
    console.log("Promise");
});

console.log("End");
```
